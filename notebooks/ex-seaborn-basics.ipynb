# Seaborn Basics

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

##### Importing data set

```python
data = pd.read_csv('./datasets/auto-mpg.csv')
data.drop(data[data['horsepower']=='?'].index, inplace = True)
data['horsepower'] = data['horsepower'].astype('float')
data['car name'] = data['car name'].astype('str')
#data['car name'].map(lambda x: type(x))
```

### 1. Pair plots

```python
# Pair plot plots a plot of all numerical columns in data frame
sns.pairplot(data)
```

```python
sliced_data = data[['model year','weight','mpg','displacement','horsepower']]
sns.pairplot(sliced_data)
```

### 3. Heatmap

```python
sns.heatmap(sliced_data.corr(), annot = True) # correlation is used for feature deduction
plt.xticks(rotation=30)
plt.title("Heat map of sliced data", fontsize = 20)
```

### 3. Box plots

```python
# box plot using seaborn
sns.boxplot(x= 'cylinders', y = 'mpg', data = data)
plt.title("Box plot of Cylinders vs MPG", fontsize = 20)
plt.xlabel("Cylinders")
plt.ylabel("MPG")
```

```python
# box plot using seaborn with hue
sns.boxplot(x= 'origin', y = 'mpg',hue = 'cylinders', data = data)
plt.title("Box plot of Cylinders vs MPG", fontsize = 20)
plt.xlabel("Orgins")
plt.ylabel("MPG")
```

### 4. Strip plots *

```python
# strip plots
sns.stripplot(x=data['mpg']) #format changes in 0.12
```

### 5. Swarm plots *

```python
# swarm plots
#plt.figure(figsize=(10,10)) if marker size is more
sns.swarmplot(x= 'cylinders',y='mpg', data = data, s=3)
```

```python
# combining in one figure
plt.figure(figsize=(8,8))
sns.stripplot(x='cylinders',y='mpg', data = data, s= 4)
sns.swarmplot(x='cylinders',y='mpg', data = data, s =4)
```

### 5. Violin plot

```python
plt.figure(figsize=(8,8))
sns.violinplot(x = 'cylinders', y = 'mpg', data = data) #box plot inside it plus rotated kernel density on each plots
# probability density smoothend by a kernel desity estimator
```

### 6. Count plots

```python
sns.countplot(x= 'origin',data = data)
plt.grid(True)
```

```python
sns.countplot(x='origin',hue='cylinders',data = data)
plt.grid(True)
```
